# Filtering

```python
filt = bikes['TRIP DURATION'] > 2500
bikes[filt].head()
```

```python
filt1 = bikes['TRIP DURATION'] > 2000
filt2 = bikes ['GENDER'] == 'Female'
filt = filt1 & filt2
bikes[filt].head()
```

```python
titanic = pd.read_csv('kapal_titanic.csv')
titanic
```

```python
# count frequency in one column

titanic['sex'].value_counts()
```

```python
titanic.groupby('sex').agg({'fare':'mean'}).round(3)
```

```python
# multiple aggregation function

titanic.groupby('sex').agg({'fare' : ['mean', 'count', 'max']}).round(3)
```

```python
# multiple grouping columns

titanic.groupby(['sex', 'age']).agg({'fare' : ['mean', 'count', 'max']}).round(3)
```

```python
kt = titanic.pivot_table(index = 'age', columns = 'sex', values = 'fare', aggfunc = 'mean').round(3)
kt
```

# Styling Pivot Table 

```python
kt.style.highlight_max()
```
