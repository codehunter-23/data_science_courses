# Boolean Indexing

```python
import pandas as pd

bikes = pd.read_csv('Divvy_Trips.csv')
bikes.head()
```

```python
bikes[['TRIP DURATION']]
```

```python
filt = bikes['TRIP DURATION'] > 1500
filt.head()
```

```python
bikes[filt].head(2)
```

```python
filt = bikes['GENDER'] == 'Female'
bikes[filt].head()
```

# Boolean Indexing Multiple Condition

```python
bikes = pd.read_csv('Divvy_Trips.csv', index_col = 'USER TYPE')
bikes.head(3)
```

```python
filt1 = bikes['TRIP DURATION'] > 2000
filt2 = bikes['GENDER'] == 'Female'
filt = filt1 & filt2

bikes[filt].head(10)
```

```python
bikes = pd.read_csv('Divvy_Trips.csv', parse_dates = ['START TIME', 'STOP TIME'])
```

```python
bikes.head(5)
```

```python
birth = bikes['BIRTH YEAR']
birth.head(3)
```

```python
filt = birth > 1990
birth[filt]
```

```python
filt1 = birth < 1975
filt2 = birth > 2000

filt = filt1 | filt2
birth[filt]
```

```python
bikes = pd.read_csv('Divvy_Trips.csv', parse_dates = ['START TIME', 'STOP TIME'], index_col = 'START TIME')
bikes.head(3)
```

```python
birth = bikes['BIRTH YEAR']
birth.head(3)
```

```python
filt = birth.between(1990,2000)
birth[filt].head()
```

```python
bikes[filt].head(5)
```
